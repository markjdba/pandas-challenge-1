## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df["line_subtotal"] = df.unit_price * df.qty

# line_sub_ship_line_price = df[["line_subtotal", "shipping_price", "line_price"]]
line_sub_ship_line_price = df[["unit_price", "qty", "line_subtotal"]]

line_sub_ship_line_price.head()

```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.

df["total_weight"] = df.unit_weight * df.qty
df['shipping_price'] =  np.where(df['total_weight'] > 50, 7*df['total_weight'], 10*df['total_weight'])
                                 
unit_price_qty_wt_ship_price = df[["unit_price", "unit_weight", "qty", "total_weight", "shipping_price"]]
unit_price_qty_wt_ship_price.head()

```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df['line_price'] = (df.shipping_price + df.line_subtotal)*1.0925 
decimals = 2 
df['line_price'] = df['line_price'].apply(lambda x: round(x, decimals))

line_sub_ship_line_price = df[["line_subtotal", "shipping_price", "line_price"]]
line_sub_ship_line_price.head() 
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df['line_cost'] = df.shipping_price + (df.unit_cost * df.qty) 
df.head() 
```

```python
# Create a column for the profit of each line using line cost and line price
df['line_profit'] = df.line_price - df.line_cost 
decimals = 2  
df['line_profit'] = df['line_profit'].apply(lambda x: round(x, decimals))

df.head() 
```

```python
# Format the data and rename the columns to names suitable for presentation.

# Define the money columns. 

# Define a function that converts a dollar amount to millions.
def currency_format_millions(x):
        intx = int(float(x))
        return format(intx/1000000)

# Apply the currency_format_millions function to only the money columns. 

summary_df['shipping_price'] = summary_df['shipping_price'].apply(currency_format_millions)
summary_df['line_price'] = summary_df['line_price'].apply(currency_format_millions)
summary_df['line_cost'] = summary_df['line_cost'].apply(currency_format_millions)
summary_df['line_profit_cost'] = summary_df['line_profit_cost'].apply(currency_format_millions)

# Rename the columns to reflect the change in the money format. 
summary_df_renamed = summary_df.rename(columns={"client_id": "Client ID",
                                                "qty": "Units",
                                                "shipping_price": "Shipping (millions)",
                                                "line_price": "Total Revenue (millions)",
                                                "line_cost": "Total Cost (millions)",
                                                "line_profit_cost": "Total Profit (millions)",
                                                })
summary_df_renamed.head()

```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
summary_df_sorted = summary_df_renamed.sort_values(by="Total Profit (millions)", ascending=False)

# Reset the index so that the index is now based on the sorting locations
summary_df_sorted = summary_df_sorted.reset_index(drop=True)

summary_df_sorted.head()
```
